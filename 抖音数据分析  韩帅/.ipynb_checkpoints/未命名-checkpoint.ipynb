## çœ‹å¤§Vå–œæ¬¢ä»€ä¹ˆç±»åˆ«çš„è§†é¢‘

```python
# 4. çœ‹å¤§Vå–œæ¬¢ä»€ä¹ˆç±»åˆ«çš„è§†é¢‘

v=df[df['custom_verify'].isnull()==False]
v.category.value_counts()[:9].plot(kind="bar",rot=0)
```

## å¤§Vçš„ç²‰ä¸æ•°çš„åˆ†å¸ƒ ä¸»è¦çš„ç²‰ä¸é‡æ˜¯å¤šå°‘

```python
# 5. å¤§Vçš„ç²‰ä¸æ•°çš„åˆ†å¸ƒ ä¸»è¦çš„ç²‰ä¸é‡æ˜¯å¤šå°‘
v['fans']
```

```python
data = pd.cut(x=v.fans.values,bins=[1000000,2000000,5000000,10000000,100000000],
              labels=['100-200ä¸‡','200-500ä¸‡','500-1000ä¸‡','1000ä¸‡ä»¥ä¸Š'])
data.value_counts().plot(kind="bar",rot=0)
```

## å‘å¸ƒæœ€å¤šçš„è§†é¢‘çš„æŠ–éŸ³è´¦å·æ˜¯å“ªäº›?TOP10

```python
# 6. å‘å¸ƒæœ€å¤šçš„è§†é¢‘çš„æŠ–éŸ³è´¦å·æ˜¯å“ªäº›?TOP10
top10 = df.sort_values(by="videos",ascending=False).iloc[:10][["name","videos"]]
top10.set_index("name").plot.bar()
```

##  æŸ¥çœ‹å“ªäº›å¤§Vè·å–çš„ç‚¹èµæ•°æ›´å¤šï¼Ÿ 

```python
# 7. æŸ¥çœ‹å“ªäº›å¤§Vè·å–çš„ç‚¹èµæ•°æ›´å¤šï¼Ÿ 
v.sort_values(by="likes",ascending=False)[["name",'likes']].set_index("name")[:15].plot.bar()
```

## æŸ¥çœ‹ç‚¹èµæ•°é‡çš„åˆ†å¸ƒ

```python
# 8. æŸ¥çœ‹ç‚¹èµæ•°é‡çš„åˆ†å¸ƒ
data = pd.cut(x=v.likes.values,bins=[1000000,2000000,5000000,10000000,100000000],
              labels=['100-200ä¸‡','200-500ä¸‡','500-1000ä¸‡','1000ä¸‡ä»¥ä¸Š'])
data.value_counts().plot(kind="bar",rot=0)
```
