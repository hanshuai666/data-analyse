#   加载数据：

```python
# 加载数据
df=pd.read_csv('./douyin.csv')
df.head()
```

## 查看大V账户的男女比例 

```python

v=df[df['custom_verify'].isnull()==False]

gender_count=v.groupby('gender')
```

```python
gender = gender_count['gender'].agg(['count'])
```

```python
gender.reset_index()
```

```python
gender.iloc[:,0]/v.shape[0]
```

```python
gender.plot.pie(subplots=True,autopct='%.1f%%')
```

## 看国外的主要网红来源有哪些国家

```python
# 2. 看国外的主要网红来源有哪些国家
df['country'].unique()
c=df.groupby('country')
```

```python
country_count=c['country'].agg(['count'])
country_count=country_count.sort_values(by='count',ascending=False)
```

## 看大V喜欢什么类别的视频

```python
# 4. 看大V喜欢什么类别的视频

v=df[df['custom_verify'].isnull()==False]
v.category.value_counts()[:9].plot(kind="bar",rot=0)
```

## 大V的粉丝数的分布 主要的粉丝量是多少

```python
# 5. 大V的粉丝数的分布 主要的粉丝量是多少
v['fans']
```

```python
data = pd.cut(x=v.fans.values,bins=[1000000,2000000,5000000,10000000,100000000],
              labels=['100-200万','200-500万','500-1000万','1000万以上'])
data.value_counts().plot(kind="bar",rot=0)
```

## 发布最多的视频的抖音账号是哪些?TOP10

```python
# 6. 发布最多的视频的抖音账号是哪些?TOP10
top10 = df.sort_values(by="videos",ascending=False).iloc[:10][["name","videos"]]
top10.set_index("name").plot.bar()
```

##  查看哪些大V获取的点赞数更多？ 

```python
# 7. 查看哪些大V获取的点赞数更多？ 
v.sort_values(by="likes",ascending=False)[["name",'likes']].set_index("name")[:15].plot.bar()
```

## 查看点赞数量的分布

```python
# 8. 查看点赞数量的分布
data = pd.cut(x=v.likes.values,bins=[1000000,2000000,5000000,10000000,100000000],
              labels=['100-200万','200-500万','500-1000万','1000万以上'])
data.value_counts().plot(kind="bar",rot=0)
```

## 查看评论、分享的分布

```python
# 9. 查看评论、分享的分布
df.groupby('city')[['comments','shares']].sum()

pd.pivot_table(df,
              index='province',
              aggfunc='sum').nlargest(15,'fans')[['comments','shares']].plot.bar()
```

##  查看大V的毕业院校主要是哪些

```python
v.groupby('school').sum().nlargest(8,'fans')[['fans']].plot.pie(subplots=True,figsize=(8,8))
```
