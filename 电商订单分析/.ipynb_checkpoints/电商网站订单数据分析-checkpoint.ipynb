### 核心指标

指标，具备业务含义，能够反映业务特征的数据，有几个要点：

1. 指标必须是数值，不能是文本、日期等。
2. 指标都是汇总计算出来的，单个明细数据不是指标。

- 新老用户
    - 新（老）用户数量占比
    - 新（老）用户金额占比
    
- 复购率和回购率
    - 复购率：复购（某段时间有2次及以上购买行为）用户的占比。复购率能反映用户的忠诚度，监测周期一般较长。
    - 回购率：回购率一般监测周期较短，可以反映如短期促销活动对用户的吸引力。
    
- 用户交易常用指标
    - 访问次数（PV）：一定时间内某个页面的浏览次数。
    - 访问人数（UV）：一定时间内访问某个页面的人数。
    - 加购数：将某款商品加入到购物车的用户数。
    - 收藏数：收藏某款商品的用户数。
    - GMV（总交易额、成交总额）：Gross Merchandise Volume，通常说的交易“流水”。
    - 客单价（ARPU）：“$ 总收入 / 总用户数 $” ---> ARPPU。
    - 转化率：“$ 付费用户数 / 访客数 $”。
    - 折扣率：“$ GMV / 吊牌总额 $”，其中吊牌总额为：“$ 吊牌价 \times 销量 $”。
    - 拒退量：拒收和退货的总数量。
    - 拒退额：拒收和退货的总金额。
    - 实际销售额：“$ GMV - 拒退额 $”。

### 数据预处理

1. 提取2019年的订单数据
2. 处理业务流程不符的数据（支付时间早于下单时间、支付时长超过30分钟、订单金额小于0、支付金额小于0）

```python
# 加载商品订单数据，id做行标签
df=pd.read_excel('./某电商网站订单数据.xlsx',index_col='id')
df.head()
```

```python
# 检查订单号有没有重复值（nunique）
df.nunique()
df['orderID'].duplicated().sum()
#备份数据
df1=df.copy()
# # 删除订单号重复的数据（保留第一项）  subset值不给，删除重复行   按指定列上删除重复记录
df.drop_duplicates(subset=['orderID'],keep='first',inplace=True)
# df
# # 提取2019年的订单数据 
# df.info()
df[df['payTime'].dt.year==2019]
# df.query("'2018-12-31 23:59:59'<payTime<'2020-01-01 00:00:00'")
# # 删除支付时间早于下单时间的数据
df[df['payTime']<df['orderTime']].index
df.drop(index=df[df['payTime']<df['orderTime']].index,inplace=True)
# # 删除支付时长超过30分钟的数据
df[df['payTime'].dt.minute-df['orderTime'].dt.minute>30]
df.drop(index=df[df['payTime'].dt.minute-df['orderTime'].dt.minute>30].index,inplace=True)
#df = df[(df.payTime-df.orderTime)<pd.to_timedelta('0 days 00:30:00')] 

# # 删除订单金额或支付金额小于0的数据
# # df.drop(index=df[df['orderAmount']<0].index,inplace=True)
# # df.drop(index=df[df['payment']<0].index,inplace=True)
# df[(df.payment<0) | (df.orderAmount<0)].index
# df.drop(index=df[(df.payment<0) | (df.orderAmount<0)].index,inplace=True)
```

### 数据清洗

3. 处理渠道为空的数据（补充众数） 离散型数据一般用众数填充
4. 处理平台类型字段（去掉多余的空格，保持数据一致）
5. 添加折扣字段，处理折扣大于1的字段（将支付金额修改为“订单金额*平均折扣”）

```python
df.isnull().any()
```

```python
df['chanelID'].fillna(df['chanelID'].mode()[0],inplace=True)
```

```python
df['chanelID'].isnull().any()
```

```python
# 将所有字符转成小写
df['platfromType'] = df['platfromType'].str.lower()
# 去掉多余的空格，保持数据一致
#r' +'正则规则，表示一个或多个空格
df['platfromType'] = df['platfromType'].str.replace(r' +','')
df2=df.copy()
```

```python
#计算折扣
df['折扣']=df['payment']/df['orderAmount']

#找出payment小于等于orderAmount的正常数据
res=df[(df.payment<=df.orderAmount)]

# 计算正常订单的平均折扣
avg_discount=res['折扣'].mean()

#将折扣大于一的数据赋值为平均折扣
res=df[df['折扣']>1]
df.loc[res.index,'折扣']=avg_discount

#根据折扣计算payment
df.loc[res.index,'payment']=df.loc[res.index,'orderAmount']*df.loc[res.index,'折扣']

```

每月GMV及趋势分析（折线图）

```python
#交易总金额（GMV）
gmv=df['orderAmount'].sum()

#总销售额
total=df['payment'].sum()

#实际销售额
real_money=df[df['chargeback']=='否'].loc[:,'payment'].sum()

#退货率
charge_back=(df[df['chargeback']=='是'].count()/df['chargeback'].count())[0]

#客单价
arppu=total/df['userID'].nunique()

print(f"交易总金额：{gmv}\n总销售额:{total}\n实际销售额:{real_money}\n退货率:{charge_back}\n客单价：{arppu}")
```

### 数据建模

- RFM模型

   - R：Recency - 最近一次购买 ---> 低
   - F: Frequency - 购买的频次 ---> 高 
   - M: Monetary - 购买总金额 ---> 高

<img src="./images/rfm-model.png" width="800">

```python
tags_dict = {
    "111":"重要价值客户",
    "101":"重要发展客户",
    "011":"重要保持客户",
    "001":"重要挽留客户",
    "110":"一般价值客户",
    "100":"一般发展客户",
    "010":"一般保持客户",
    "000":"一般挽留客户"
}
```

```python
#增加一列
df['F']=1
# rfm=pd.pivot_table(df,index='userID',
#                    aggfunc={'payTime':'max','F':'sum','payment':'sum'})

rfm=df.groupby(by='userID').agg({'payTime':'max','F':'sum','payment':'sum'})
```

```python
from datetime import datetime
```

```python
end=datetime(2019,12,31,23,59,59)
```

```python
#计算购买日期到2019年低的间隔天数
rfm['payTime']=(end-rfm['payTime']).dt.days
```

```python
rfm.columns=['R','F','M']
rfm
```

```python
#划分标准：根据平均值划分
#rfm.apply(lambda x:print(x,end='\n\n'),axis=0)
r,f,m=rfm['R'].mean(),rfm['F'].mean(),rfm['M'].mean()
r,f,m
```

```python
#列的转换
a=rfm['R'].apply(lambda x:'1' if x>r else '0')
```

```python
#列的转换
b=rfm['F'].apply(lambda x:'1' if x>f else '0')
```

```python
#列的转换
c=rfm['M'].apply(lambda x:'1' if x>m else '0')
```

```python
rfm['tag']=a+b+c
```

```python
rfm
```

```python
#转换标签列
rfm['tag']=rfm['tag'].map({
    "111":"重要价值客户",
    "101":"重要发展客户",
    "011":"重要保持客户",
    "001":"重要挽留客户",
    "110":"一般价值客户",
    "100":"一般发展客户",
    "010":"一般保持客户",
    "000":"一般挽留客户"
})
```
